**a) Пример простого диалога в консоли:**

Ключ: [40, 20, 90, 10] (данные в процентах)
Найти замок?

1: [60, 80, 10, 90]

2: [60, 70, 20, 90]

3: [50, 80, 10, 20]

4: [40, 50, 30, 10]


```python
using System;

class GameFindLock
{
    static void Main()
    {
        // Ключ
        int[] key = { 40, 20, 90, 10 };

        // Список замков
        int[][] locks = new int[][]
        {
            new int[] { 60, 80, 10, 90 },
            new int[] { 60, 70, 20, 90 },
            new int[] { 50, 80, 10, 20 },
            new int[] { 40, 50, 30, 10 }
        };

        // Выводим ключ
        Console.WriteLine("Ключ: [" + string.Join(", ", key) + "]");
        Console.WriteLine("\nНайти замок?");

        // Выводим замки
        for (int i = 0; i < locks.Length; i++)
        {
            Console.WriteLine($"{i + 1}: [{string.Join(", ", locks[i])}]");
        }

        Console.Write("\nВведите номер замка: ");
        if (int.TryParse(Console.ReadLine(), out int choice) && choice >= 1 && choice <= locks.Length)
        {
            int[] selectedLock = locks[choice - 1];
            bool isMatch = IsMatch(key, selectedLock, tolerance: 10);

            Console.WriteLine(isMatch
                ? "✅ Замок подходит!"
                : "❌ Замок не подходит.");
        }
        else
        {
            Console.WriteLine("Неверный ввод.");
        }
    }

    // Сравнение замка с ключом с допуском
    static bool IsMatch(int[] key, int[] test, int tolerance)
    {
        if (key.Length != test.Length) return false;

        for (int i = 0; i < key.Length; i++)
        {
            if (Math.Abs(key[i] - test[i]) > tolerance)
                return false;
        }
        return true;
    }
}
```
